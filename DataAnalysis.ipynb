```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# read in the CSV
df = pd.read_csv("./Resources/cardio_train.csv", sep = ";", index_col=0)
df.head()
```

```python
# Adding a BMI column
height_meter = df["height"]/100
df["BMI"] = df["weight"]/height_meter/height_meter
df.head()
```

```python
# plotting the BMI vs cardio disease
df.plot(x = "BMI", y = "cardio", kind = "scatter")
```

```python
# examine why there are some high BMI values
highBMIdf = df[df["BMI"] > 50]
highBMIdf.head()
```

```python
# Plotting the histogram
df_cardio = df[df["cardio"] == 0]
df_n_cardio = df[df["cardio"] == 1]

plt.hist([df_cardio["height"], df_n_cardio["height"]], color = ["Black", "Red"], label = ["w_cardio", "wo_cardio"], bins = 50)
plt.xlabel("Height")
plt.title("Height Distribution")
plt.legend()

plt.show()
```
